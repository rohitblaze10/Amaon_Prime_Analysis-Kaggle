## EDA -Exploratory Data Analysis

```python
df.describe()
```

```python
df["Subscription Plan"].value_counts()
```

```python
sns.histplot(df["Age"],bins=20,kde=True)
plt.title("Distribution of Users")
plt.show()
```

```python
sns.countplot(x="Subscription Plan",data=df)
plt.title("Subsription Plan distribution")
plt.show()
```

```python
sns.scatterplot(x="Engagement Metrics",y="Feedback/Ratings",data=df,alpha=0.5)
plt.title("Engagement v/s Ratings")
plt.show()
```

```python
device_counts = df['Devices Used'].value_counts()


sns.barplot(x=device_counts.index, y=device_counts.values, palette='Purples')
plt.title("Device Usage for Amazon Prime")
plt.xlabel("Device")
plt.ylabel("Number of Users")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.set_style("whitegrid")
plt.show()
```

```python
df["Age Group"]= pd.cut(df['Age'], bins=[0, 18, 30, 45, 60, 100], 
                         labels=['Under 18', '18-30', '30-45', '45-60', '60+'])

plt.figure(figsize=(8, 5))
age_group_counts = df['Age Group'].value_counts()
sns.barplot(x=age_group_counts.index, y=age_group_counts.values, palette='coolwarm')
plt.title("Age Group Distribution of Amazon Prime Users")
plt.xlabel("Age Group")
plt.ylabel("Number of Users")
plt.show(10)



```
